```python
import click
import logging
from pathlib import Path
from GeoUtilities import DataConverter
import os, json, shutil

import glob
from sklearn.model_selection import train_test_split
```

```python
data_name = "Vegas_v1"

main_dir = "/home/hgamarro/DeepLearning/HG_space/src/data"
project_dir = Path(main_dir).resolve().parents[1]

interim_dir = str(project_dir) + '/data/interim/' 
processed_dir = str(project_dir) + '/data/processed/' 
```

```python
SN_idx = 2
SN_name = "Vegas"
SN_input_path = str(project_dir) + '/data/raw/' + "AOI_{}_{}_Train/".format(SN_idx,SN_name )

SN_interim_train_path = interim_dir + data_name +'/train'
SN_interim_val_path = interim_dir + data_name + '/val'
SN_interim_test_path = interim_dir + data_name + '/test'

SN_processed_train_path = processed_dir + data_name +'/train'
SN_processed_val_path = processed_dir + data_name + '/val'
SN_processed_test_path = processed_dir + data_name + '/test'
```

```python
if data_name not in os.listdir(interim_dir): 
    os.makedirs(interim_dir + data_name + "/train/images", exist_ok=True)
    os.makedirs(interim_dir + data_name + "/train/masks", exist_ok=True)
    os.makedirs(interim_dir + data_name + "/val/images", exist_ok=True)
    os.makedirs(interim_dir + data_name + "/val/masks", exist_ok=True)
    os.makedirs(interim_dir + data_name + "/test/images", exist_ok=True)
    os.makedirs(interim_dir + data_name + "/test/masks", exist_ok=True)

if data_name not in os.listdir(processed_dir):
    os.makedirs(processed_dir + data_name + "/train/images", exist_ok=True)
    os.makedirs(processed_dir + data_name + "/train/masks", exist_ok=True)
    os.makedirs(processed_dir + data_name + "/val/images", exist_ok=True)
    os.makedirs(processed_dir + data_name + "/val/masks", exist_ok=True)
    os.makedirs(processed_dir + data_name + "/test/images", exist_ok=True)
    os.makedirs(processed_dir + data_name + "/test/masks", exist_ok=True)
```

```python
converter = DataConverter.SpaceNetDataConverter(SN_input_path, SN_interim_train_path, SN_idx, SN_name)
converter.convertAllToInput()
```

```python
percent_val = int(len(glob.glob(SN_interim_train_path + "/images" + '/*.png')) *.75) / len(glob.glob(SN_interim_train_path + "/images" + '/*.png'))
DataConverter.train_val_split(SN_interim_train_path, SN_interim_val_path, percent_val)

percent_test = int(len(glob.glob(SN_interim_train_path + "/images" + '/*.png')) *.85) / len(glob.glob(SN_interim_train_path + "/images" + '/*.png'))
DataConverter.train_val_split(SN_interim_train_path, SN_interim_test_path, percent_test)

total_imgs = len(glob.glob(SN_interim_train_path + "/images" + '/*.png')) + len(glob.glob(SN_interim_val_path + "/images" + '/*.png')) + len(glob.glob(SN_interim_test_path + "/images" + '/*.png'))
print(len(glob.glob(SN_interim_val_path + "/images" + '/*.png')) / total_imgs)
print(len(glob.glob(SN_interim_test_path + "/images" + '/*.png')) / total_imgs)

```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img_list_train = sorted(glob.glob(SN_interim_train_path + "/images" + '/*.png'))
mask_list_train = sorted(glob.glob(SN_interim_train_path + "/masks" + '/*.png'))


fig, axes = plt.subplots(2, 8, figsize=(16, 5))
# fig.subplots_adjust(hspace=0.1)

for a in np.arange(0,8):
    
    
    idx_img = a+55
    data_img = plt.imread(img_list_train[idx_img])
    data_mask =plt.imread(mask_list_train[idx_img])

    axes[0, a].imshow( data_img)
    axes[1, a].imshow( data_mask)
```
